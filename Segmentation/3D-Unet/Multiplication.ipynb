```python
import nibabel
import numpy
import matplotlib.pyplot as plt
```

```python
image = nibabel.load ("E:/Data/Dataset3/images/colon_219.nii.gz").get_fdata().astype(int)
mask = nibabel.load ("E:/Data/Dataset3/seg_219.nii").get_fdata().astype(int)
```

```python
image.shape
```

```python
mask.shape
```

```python
image_vec = numpy.reshape(image, (image.shape[0]*image.shape[1]*image.shape[2],))
```

```python
mask_vec = numpy.reshape(mask, (mask.shape[0]*mask.shape[1]*mask.shape[2],))
```

```python
image_vec.shape
```

```python
mask_vec.shape
```

```python
multi = numpy.multiply (image_vec, mask_vec)
```

```python
multi.shape
```

```python
tensor = numpy.reshape (multi, (512,512,137) )
```

```python
tensor.shape
```

```python
plt.imshow(tensor[:, :, 40])
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(figsize=(18, 6), ncols=3)
ax1.title.set_text('Image')
ax2.title.set_text('Predicted Mask')
ax3.title.set_text('Segmented Image')
bar1 = ax1.imshow(image [ :, :, 40], cmap="gray")
fig.colorbar(bar1, ax=ax1)
bar2 = ax2.imshow(mask [ :, :, 40],  cmap="jet")
fig.colorbar(bar2, ax=ax2)
bar3 = ax3.imshow(tensor [ :, :, 40], cmap="jet")
fig.colorbar(bar3, ax=ax3)
plt.show()
```
